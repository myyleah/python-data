## Exploratory Data Analysis for all Data Roles

### Roles to Explore

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# Prepare the data
df_plot = df['job_title_short'].value_counts().to_frame().reset_index()
df_plot.columns = ['job_title_short', 'count']

# Create a custom pink gradient colormap
pink_cmap = LinearSegmentedColormap.from_list('pink_gradient', ['#C71585', '#FFC0CB'])  # dark pink to light pink

# Create the pink palette based on number of job titles
pink_palette = sns.color_palette(pink_cmap(np.linspace(0, 1, len(df_plot))))

# Plot
sns.set_theme(style='ticks')
sns.barplot(
    data=df_plot,
    x='count',
    y='job_title_short',
    hue='count',                  # still using count to color
    palette=pink_palette,
    legend=False
)
sns.despine()

plt.title('Number of Jobs per Job Title')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

### Countries to Explore

```python
# Prepare the data
df_plot = df['job_country'].value_counts().to_frame().reset_index().head(20)
df_plot.columns = ['job_country', 'count']

# Create a custom pink gradient colormap
pink_cmap = LinearSegmentedColormap.from_list('pink_gradient', ['#C71585', '#FFC0CB'])  # dark pink to light pink

# Create the pink palette based on number of countries
pink_palette = sns.color_palette(pink_cmap(np.linspace(0, 1, len(df_plot))))

# Plot
sns.set_theme(style='ticks')
sns.barplot(
    data=df_plot,
    x='count',
    y='job_country',
    hue='count',
    palette=pink_palette,
    legend=False
)
sns.despine()

plt.title('Number of Jobs per Country')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

### Companies to Explore

```python
# Prepare the data
df_plot = df['company_name'].value_counts().to_frame()[1:].reset_index().head(20)
df_plot.columns = ['company_name', 'count']

# Create a custom pink gradient colormap
pink_cmap = LinearSegmentedColormap.from_list('pink_gradient', ['#C71585', '#FFC0CB'])  # dark pink to light pink

# Create the pink palette based on number of companies
pink_palette = sns.color_palette(pink_cmap(np.linspace(0, 1, len(df_plot))))

# Plot
sns.set_theme(style='ticks')
sns.barplot(
    data=df_plot,
    x='count',
    y='company_name',
    hue='count',
    palette=pink_palette,
    legend=False
)
sns.despine()

plt.title('Number of Jobs per Company')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

### Job Opportunities

```python
dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

# Define cute pink colors
colors = ['#FFB6C1', '#FFC0CB']  # Light pink shades

fig, ax = plt.subplots(1, 3, figsize=(11, 3.5))

for i, (column, title) in enumerate(dict_column.items()):
    counts = df[column].value_counts()
    ax[i].pie(
        counts,
        labels=['False', 'True'],
        autopct='%1.1f%%',
        startangle=90,
        colors=colors   # <-- pink colors added here
    )
    ax[i].set_title(title)

plt.tight_layout()
plt.show()
```

## Exploratory Data Analysis for Data Analysts in the US

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

### Filter for US Data Analyst roles

```python
df_DA_US = df[(df['job_country'] == 'United States') & (df['job_title_short'] == 'Data Analyst')]
```

### Locations to Explore:

```python

# Prepare the data
df_plot = df_DA_US['job_location'].value_counts().head(10).to_frame().reset_index()
df_plot.columns = ['job_location', 'count']

# Create a custom pink gradient colormap
pink_cmap = LinearSegmentedColormap.from_list('pink_gradient', ['#C71585', '#FFC0CB'])  # dark pink to light pink

# Create the pink palette based on number of job locations
pink_palette = sns.color_palette(pink_cmap(np.linspace(0, 1, len(df_plot))))

# Plot
sns.set_theme(style='ticks')
sns.barplot(
    data=df_plot,
    x='count',
    y='job_location',
    hue='count',
    palette=pink_palette,
    legend=False
)
sns.despine()

plt.title('Counts of Job Locations for Data Analyst in the US')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```
